```python
def arithmetic_arranger(string_form, state_ment):
    
    # Considering the input as a list:
    list_string_form = string_form
    
    # Limiting problems up to 5, anything more will return "Error: Too many problems."
    if len(list_string_form) >= 5:
        return ("Error: Too many problems.")

    #String cache for print format in horizontal arrangement.
    first_row_setup = ""
    second_row_setup = ""
    third_row_setup = ""
    fourth_row_setup = ""

    # Looping for each problem:
    for string_form in list_string_form:
        
        # Splitting each problem by a " ":
        split_string_form = string_form.split(" ")

        # Assigning strings for operands and operators:
        str_first_number = split_string_form[0]
        str_operator = split_string_form[1]
        str_second_number = split_string_form[2]
        
        # Making sure each problem comprises only of digits, otherwise 
        # "Error: Numbers must only contain digits." will be returned.
        if str_first_number.isdigit() != True or str_second_number.isdigit() != True:
            return ("Error: Numbers must only contain digits.")
        
        # Making sure the operator comprises only of '+' or '-', otherwise
        # "Error: Operator must be '+' or '-'." will be returned.
        if str_operator != '+':
            if str_operator != '-':
                return ("Error: Operator must be '+' or '-'.")
        
        # Limiting the number of digits of operands up to 4 only, otherwise
        # "Error: Numbers cannot be more than four digits." will be returned.
        if len(str_first_number) > 4 or len(str_second_number) > 4:
            return ("Error: Numbers cannot be more than four digits.")
        
        # Conditions were based on the differences between the number of digits of operands:
        # Both operands have the same number of digits:
        elif abs((len(str_first_number) - len(str_second_number))) == 0:
            
            # There must be a single pace between the operator and the longest of the two operands
            # Assigning a white_space variable defining the maximum space required to cover the problem.
            white_space = len(str_first_number) + 2
            
            # Numbers should be right-aligned.
            rjust_1n = str_first_number.rjust(white_space)
            
            # In this case, the white_space of the second number will be reduced by 1 since --
            # -- the operator occupies the leftmost space.
            # There must be a single pace between the operator and the longest of the two operands
            rjust_2n = str_second_number.rjust(white_space-1)
            
            # The operand placed on the leftmost space of the problem.
            ljust_op = str_operator.ljust(0)

            # Condition for the function if the second argument is set to "True".
            if state_ment == True:
                
                # The answers of the problems in the first argument will be displayed.
                if ljust_op == '+':
                    rjust_answer = str(int(str_first_number)+int(str_second_number)).rjust(white_space)
                elif ljust_op == '-':
                    rjust_answer = str(int(str_first_number)-int(str_second_number)).rjust(white_space)
            
            # If set to false, answers will not be displayed.
            else:
                rjust_answer = ""

            # Assigning variables for the arrangement of the problem format per row.
            # Both operands will be in the same order as provided --  
            # -- (the first will be the top one and the second will be the bottom.)
            first_row = rjust_1n
            
            # the operator will be on the same line as the second operand
            second_row = ljust_op + rjust_2n
            
            # There should be dashes at the bottom of each problem. 
            # The dashes should run along the entire length of each problem individually.
            third_row = white_space*"-"
            
            # Answer to the problem placed on the last row
            fourth_row = rjust_answer
            
            # Adding the new variables to the print format per row.
            first_row_setup += first_row + "    " # Adding four spaces between each problem.
            second_row_setup += second_row + "    " # Adding four spaces between each problem.
            third_row_setup += third_row + "    " # Adding four spaces between each problem.
            fourth_row_setup += fourth_row + "    " # Adding four spaces between each problem.

        # The operands have a 1 digit difference:
        elif abs((len(str_first_number) - len(str_second_number))) == 1:
            
            # The maximum number of digits for an operand is set to 4, which means that the maximum -- 
            # -- white space is 6, even for (3,4) digits.
            if len(str_first_number) + len(str_second_number) >= 6:
                white_space = 6
                
            # Special conditions covering the (1,2) and (2,3) digits:
            elif len(str_first_number) + len(str_second_number) < 6:
                
                # The maximum white space for the (1,2) digit condition is 4:
                if len(str_first_number) + len(str_second_number) == 3:
                    white_space = 4
                
                # The maximum white space for the (2,3) digit condition is 5 or the number of digits -- 
                # -- of both operands combined.
                else:
                    white_space = len(str_first_number) + len(str_second_number)

            # Numbers should be right-aligned.
            rjust_1n = str_first_number.rjust(white_space)
            
            # In this case, the white_space of the second number will be reduced by 1 since --
            # -- the operator occupies the leftmost space.
            # There must be a single pace between the operator and the longest of the two operands
            rjust_2n = str_second_number.rjust(white_space-1)
            
            # The operand placed on the leftmost space of the problem.
            ljust_op = str_operator.ljust(0)

            # Condition for the function if the second argument is set to "True".
            if state_ment == True:
                
                # The answers of the problems in the first argument will be displayed.
                if ljust_op == '+':
                    rjust_answer = str(int(str_first_number)+int(str_second_number)).rjust(white_space)
                elif ljust_op == '-':
                    rjust_answer = str(int(str_first_number)-int(str_second_number)).rjust(white_space)
            
            # If set to false, answers will not be displayed.
            else:
                rjust_answer = ""

            # Assigning variables for the arrangement of the problem format per row.
            # Both operands will be in the same order as provided --  
            # -- (the first will be the top one and the second will be the bottom.)
            first_row = rjust_1n
            
            # the operator will be on the same line as the second operand
            second_row = ljust_op + rjust_2n
            
            # There should be dashes at the bottom of each problem. 
            # The dashes should run along the entire length of each problem individually.
            third_row = white_space*"-"
            
            # Answer to the problem placed on the last row
            fourth_row = rjust_answer
            
            # Adding the new variables to the print format per row.
            first_row_setup += first_row + "    " # Adding four spaces between each problem.
            second_row_setup += second_row + "    " # Adding four spaces between each problem.
            third_row_setup += third_row + "    " # Adding four spaces between each problem.
            fourth_row_setup += fourth_row + "    " # Adding four spaces between each problem.
            
        # The operands have a 2 digit difference:
        elif abs((len(str_first_number) - len(str_second_number))) == 2:
            
            # Special condition covering the (2,4) digits:
            # Since a number with 4 digits is the longest digit, the white space is set to 6.
            if len(str_first_number) + len(str_second_number) >= 6:
                white_space = 6
                
            # Special conditions covering the (1,3) digits:
            # Since a number with 3 digits is the longest digit, the white space is set to 5 -- 
            # -- or the number of digits in both operands compoined + 1
            elif len(str_first_number) + len(str_second_number) < 6:
                white_space = len(str_first_number) + len(str_second_number) + 1
                
            rjust_1n = str_first_number.rjust(white_space)
            rjust_2n = str_second_number.rjust(white_space-1)
            ljust_op = str_operator.ljust(0)

            if state_ment == True:
                if ljust_op == '+':
                    rjust_answer = str(int(str_first_number)+int(str_second_number)).rjust(white_space)
                elif ljust_op == '-':
                    rjust_answer = str(int(str_first_number)-int(str_second_number)).rjust(white_space)
            else:
                rjust_answer = ""
                
            # Numbers should be right-aligned.
            rjust_1n = str_first_number.rjust(white_space)
            
            # In this case, the white_space of the second number will be reduced by 1 since --
            # -- the operator occupies the leftmost space.
            # There must be a single pace between the operator and the longest of the two operands
            rjust_2n = str_second_number.rjust(white_space-1)
            
            # The operand placed on the leftmost space of the problem.
            ljust_op = str_operator.ljust(0)

            # Condition for the function if the second argument is set to "True".
            if state_ment == True:
                
                # The answers of the problems in the first argument will be displayed.
                if ljust_op == '+':
                    rjust_answer = str(int(str_first_number)+int(str_second_number)).rjust(white_space)
                elif ljust_op == '-':
                    rjust_answer = str(int(str_first_number)-int(str_second_number)).rjust(white_space)
            
            # If set to false, answers will not be displayed.
            else:
                rjust_answer = ""

            # Assigning variables for the arrangement of the problem format per row.
            # Both operands will be in the same order as provided --  
            # -- (the first will be the top one and the second will be the bottom.)
            first_row = rjust_1n
            
            # the operator will be on the same line as the second operand
            second_row = ljust_op + rjust_2n
            
            # There should be dashes at the bottom of each problem. 
            # The dashes should run along the entire length of each problem individually.
            third_row = white_space*"-"
            
            # Answer to the problem placed on the last row
            fourth_row = rjust_answer
            
            # Adding the new variables to the print format per row.
            first_row_setup += first_row + "    " # Adding four spaces between each problem.
            second_row_setup += second_row + "    " # Adding four spaces between each problem.
            third_row_setup += third_row + "    " # Adding four spaces between each problem.
            fourth_row_setup += fourth_row + "    " # Adding four spaces between each problem.
            
        # The operands have a 3 digit difference:
        elif abs((len(str_first_number) - len(str_second_number))) == 3:
            
            # Since a number with 3 digits is the longest digit, the white space is set to 5 -- 
            # -- or the number of digits in both operands compoined + 1
            white_space = len(str_first_number) + len(str_second_number) +1
            
            
            # Numbers should be right-aligned.
            rjust_1n = str_first_number.rjust(white_space)
            
            # In this case, the white_space of the second number will be reduced by 1 since --
            # -- the operator occupies the leftmost space.
            # There must be a single pace between the operator and the longest of the two operands
            rjust_2n = str_second_number.rjust(white_space-1)
            
            # The operand placed on the leftmost space of the problem.
            ljust_op = str_operator.ljust(0)

            # Condition for the function if the second argument is set to "True".
            if state_ment == True:
                
                # The answers of the problems in the first argument will be displayed.
                if ljust_op == '+':
                    rjust_answer = str(int(str_first_number)+int(str_second_number)).rjust(white_space)
                elif ljust_op == '-':
                    rjust_answer = str(int(str_first_number)-int(str_second_number)).rjust(white_space)
            
            # If set to false, answers will not be displayed.
            else:
                rjust_answer = ""

            # Assigning variables for the arrangement of the problem format per row.
            # Both operands will be in the same order as provided --  
            # -- (the first will be the top one and the second will be the bottom.)
            first_row = rjust_1n
            
            # the operator will be on the same line as the second operand
            second_row = ljust_op + rjust_2n
            
            # There should be dashes at the bottom of each problem. 
            # The dashes should run along the entire length of each problem individually.
            third_row = white_space*"-"
            
            # Answer to the problem placed on the last row
            fourth_row = rjust_answer
            
            # Adding the new variables to the print format per row.
            first_row_setup += first_row + "    " # Adding four spaces between each problem.
            second_row_setup += second_row + "    " # Adding four spaces between each problem.
            third_row_setup += third_row + "    " # Adding four spaces between each problem.
            fourth_row_setup += fourth_row + "    " # Adding four spaces between each problem.

    print(first_row_setup) # Printing the string caches according to row.
    print(second_row_setup) # Printing the string caches according to row.
    print(third_row_setup) # Printing the string caches according to row.
    print(fourth_row_setup) # Printing the string caches according to row.
```
