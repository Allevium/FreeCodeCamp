```python
class Category:
        
    def __init__(self, category):
        
        # Assigning the indicated category to
        self.category = category
        
        # Making lists for ledger, total balance, and their values
        self.food_led = []
        self.food_tb = 0
        self.food_led_desc_only = []
        self.food_tb_values = []
        
        # Making the withdraw values global for later use
        global food_withlist
        food_withlist = []
        
        # Making lists for ledger, total balance, and their values
        self.ent_led = []
        self.ent_tb = 0
        self.ent_led_desc_only = []
        self.ent_tb_values = []
        
        # Making the withdraw values global for later use
        global ent_withlist
        ent_withlist = [] 
        
        # Making lists for ledger, total balance, and their values
        self.cl_led = []
        self.cl_tb = 0
        self.cl_led_desc_only = []
        self.cl_tb_values = []
        
        # Making the withdraw values global for later use
        global cl_withlist 
        cl_withlist = []
        
        # Making the withdraw values global for later use
        global extra_withlist
        extra_withlist = []
        
        # Making lists for ledger, total balance, and their values
        self.ledger = []
        self.total_balance = 0
        self.ledger_desc_only = []
        self.total_balance_values = []
    
    def __str__(self):
        
        # Making variable strings to follow format
        # Asterisk characters required for first line:
        char = '*'
        
        # For the first line:
        title = ''
        
        # For the last line:
        total_ = "Total: "
        
        # For returning the budget list format
        budget_list_format = ""
        
        # If category is food:
        if self.category == "Food":
            cat_name = "Food"
            
            # Placing the word 'Food' in a center of 30 characters.
            for i in range(13):
                title += char

            title += cat_name

            for i in range(13):
                title += char
            print(title)
            
            # Looping through each transaction:
            n = 0
            for price_ in self.food_tb_values:

                # Recording the number of characters of each price:
                price_len = len(str(self.food_tb_values[n]))
                
                # Recording the number of characters of each description:
                desc_len = len(self.food_led_desc_only[n])
                
                # If the price is a whole number:
                if price_/ int(price_) == 1:
                    
                    # Assigning the value to a string to apply rjust:
                    str_price_list = str(self.food_tb_values[n])
                    
                    # If the number of chars of the description is longer than the -- 
                    # -- the number of chars of price and 3 additional chars (.00):
                    if len(self.food_led_desc_only[n]) > 27 - price_len:
                        
                        # Assigining the position of rjust:
                        text_limit = 27 - price_len
                        
                        # Left justifying the description:
                        desc_adjust = self.food_led_desc_only[n].ljust(0)
                        
                        # Cutting the description up to a space before it overlaps with -- 
                        # -- the price chars:
                        desc_adjust = desc_adjust[0:text_limit - 1]
                        
                        # Right justifying the price chars:
                        price_adjust = str_price_list.rjust(price_len)
                        
                        # Adding the line to the format:
                        budget_list_format += (desc_adjust + " " + price_adjust + ".00\n")
                        
                        # Moving on to the next iteration
                        n = n + 1
                    
                    # If the price and description chars won't overlap:
                    else:
                        text_limit = 27 - desc_len
                        desc_adjust = self.food_led_desc_only[n].ljust(0)
                        price_adjust = str_price_list.rjust(text_limit)
                        budget_list_format +=(desc_adjust + price_adjust + ".00\n")
                        n = n + 1 
                        
                #If the price has decimals:            
                else:
                    
                    # Metric for determining the number of decimals:
                    dec_places = len(str(round(abs(price_) - int(abs(price_)),2)))

                    # If the number of decimals is 1:
                    if dec_places == 3:
                        
                        #Repeat earlier process
                        str_price_list = str(self.food_tb_values[n])
                        if len(self.food_led_desc_only[n]) > 29 - price_len:
                            text_limit = 29 - price_len
                            desc_adjust = self.food_led_desc_only[n].ljust(0)
                            desc_adjust = desc_adjust[0:text_limit - 1]
                            price_adjust = str_price_list.rjust(price_len)
                            budget_list_format += (desc_adjust + " " + price_adjust + "0\n")
                            n = n + 1
                        else:
                            text_limit = 29 -  desc_len
                            desc_adjust = self.food_led_desc_only[n].ljust(0)
                            price_adjust = str_price_list.rjust(text_limit)
                            budget_list_format += (desc_adjust + price_adjust + "0\n")
                            n = n + 1    
    
                    # If the number of decimals is 2:
                    elif dec_places == 4:
                
                        # Repeat the earlier process
                        str_price_list = str(self.food_tb_values[n])
                        if len(self.food_led_desc_only[n]) > 30 - price_len:
                            text_limit = 30 - price_len
                            desc_adjust = self.food_led_desc_only[n].ljust(0)
                            desc_adjust = desc_adjust[0:text_limit - 1]
                            price_adjust = str_price_list.rjust(price_len)
                            budget_list_format +=(desc_adjust + " " + price_adjust + "\n")
                            n = n + 1
                        else:
                            text_limit = 30 - desc_len
                            desc_adjust = self.food_led_desc_only[n].ljust(0)
                            price_adjust = str_price_list.rjust(text_limit)
                            budget_list_format += (desc_adjust + price_adjust + "\n")
                            n = n + 1    
                            
            # Consolidating the lines into a whole string:
            budget_list_format += (total_ + str(self.food_tb))
            
            # Returning the string when printed:
            return budget_list_format
        
        # Repeating the process for this category:
        elif self.category == "Entertainment":
            cat_name = "Entertainment"

            for i in range(6):
                title += char

            title += cat_name

            for i in range(7):
                title += char
            print(title)
            
            n = 0
            for price_ in self.ent_tb_values:


                price_len = len(str(self.ent_tb_values[n]))
                desc_len = len(self.ent_led_desc_only[n])

                if price_/ int(price_) == 1:
                    str_price_list = str(self.ent_tb_values[n])
                    if len(self.ent_led_desc_only[n]) > 27 - price_len:
                        text_limit = 27 - price_len
                        desc_adjust = self.ent_led_desc_only[n].ljust(0)
                        desc_adjust = desc_adjust[0:text_limit - 1]
                        price_adjust = str_price_list.rjust(price_len)
                        budget_list_format += (desc_adjust + " " + price_adjust + ".00\n")
                        n = n + 1
                    else:
                        text_limit = 27 - desc_len
                        desc_adjust = self.ent_led_desc_only[n].ljust(0)
                        price_adjust = str_price_list.rjust(text_limit)
                        budget_list_format +=(desc_adjust + price_adjust + ".00\n")
                        n = n + 1    

                else:
                    dec_places = len(str(round(abs(price_) - int(abs(price_)),2)))

                    if dec_places == 3:
                        str_price_list = str(self.ent_tb_values[n])
                        if len(self.ent_led_desc_only[n]) > 29 - price_len:
                            text_limit = 29 - price_len
                            desc_adjust = self.ent_led_desc_only[n].ljust(0)
                            desc_adjust = desc_adjust[0:text_limit - 1]
                            price_adjust = str_price_list.rjust(price_len)
                            budget_list_format +=(desc_adjust + " " + price_adjust + "0\n")
                            n = n + 1
                        else:
                            text_limit = 29 -  desc_len
                            desc_adjust = self.ent_led_desc_only[n].ljust(0)
                            price_adjust = str_price_list.rjust(text_limit)
                            budget_list_format +=(desc_adjust + price_adjust + "0\n")
                            n = n + 1    

                    elif dec_places == 4:
                        str_price_list = str(self.ent_tb_values[n])
                        if len(self.ent_led_desc_only[n]) > 30 - price_len:
                            text_limit = 30 - price_len
                            desc_adjust = self.ent_led_desc_only[n].ljust(0)
                            desc_adjust = desc_adjust[0:text_limit - 1]
                            price_adjust = str_price_list.rjust(price_len)
                            budget_list_format +=(desc_adjust + " " + price_adjust + "\n")
                            n = n + 1
                        else:
                            text_limit = 30 - desc_len
                            desc_adjust = self.ent_led_desc_only[n].ljust(0)
                            price_adjust = str_price_list.rjust(text_limit)
                            budget_list_format +=(desc_adjust + price_adjust + "\n")
                            n = n + 1    

            budget_list_format +=(total_ + str(self.ent_tb))
            return budget_list_format
            
        # Repeating the process for this category:    
        elif self.category == "Clothing":
            cat_name = "Clothing"

            for i in range(11):
                title += char

            title += cat_name

            for i in range(11):
                title += char
            print(title)
            
            n = 0
            for price_ in self.cl_tb_values:


                price_len = len(str(self.cl_tb_values[n]))
                desc_len = len(self.cl_led_desc_only[n])

                if price_/ int(price_) == 1:
                    str_price_list = str(self.cl_tb_values[n])
                    if len(self.cl_led_desc_only[n]) > 27 - price_len:
                        text_limit = 27 - price_len
                        desc_adjust = self.cl_led_desc_only[n].ljust(0)
                        desc_adjust = desc_adjust[0:text_limit - 1]
                        price_adjust = str_price_list.rjust(price_len)
                        budget_list_format +=(desc_adjust + " " + price_adjust + ".00\n")
                        n = n + 1
                    else:
                        text_limit = 27 - desc_len
                        desc_adjust = self.cl_led_desc_only[n].ljust(0)
                        price_adjust = str_price_list.rjust(text_limit)
                        budget_list_format +=(desc_adjust + price_adjust + ".00\n")
                        n = n + 1    
                else:
                    dec_places = len(str(round(abs(price_) - int(abs(price_)),2)))

                    if dec_places == 3:
                        str_price_list = str(self.cl_tb_values[n])
                        if len(self.cl_led_desc_only[n]) > 29 - price_len:
                            text_limit = 29 - price_len
                            desc_adjust = self.cl_led_desc_only[n].ljust(0)
                            desc_adjust = desc_adjust[0:text_limit - 1]
                            price_adjust = str_price_list.rjust(price_len)
                            budget_list_format +=(desc_adjust + " " + price_adjust + "0\n")
                            n = n + 1
                        else:
                            text_limit = 29 -  desc_len
                            desc_adjust = self.cl_led_desc_only[n].ljust(0)
                            price_adjust = str_price_list.rjust(text_limit)
                            budget_list_format +=(desc_adjust + price_adjust + "0\n")
                            n = n + 1    

                    elif dec_places == 4:
                        str_price_list = str(self.cl_tb_values[n])
                        if len(self.cl_led_desc_only[n]) > 30 - price_len:
                            text_limit = 30 - price_len
                            desc_adjust = self.cl_led_desc_only[n].ljust(0)
                            desc_adjust = desc_adjust[0:text_limit - 1]
                            price_adjust = str_price_list.rjust(price_len)
                            budget_list_format +=(desc_adjust + " " + price_adjust + "\n")
                            n = n + 1
                        else:
                            text_limit = 30 - desc_len
                            desc_adjust = self.cl_led_desc_only[n].ljust(0)
                            price_adjust = str_price_list.rjust(text_limit)
                            budget_list_format +=(desc_adjust + price_adjust + "\n")
                            n = n + 1    

            budget_list_format +=(total_ + str(self.cl_tb))
            return budget_list_format
            
    # def deposit: If no description is given, it should default to an empty string. 
    # The method should append an object to the ledger list in the form of --
    # -- {"amount": amount, "description": description}.
    def deposit(self, amount, description = None):
        
        # If there is no description, a blank is defined.
        if description == None:
            description = ''
        
        # Assigning the amount and description into a dictionary:
        dep_details = {"amount": amount, "description": description}
        
        # Appending the details when a value is deposited into the food cat:
        if self.category == "Food":
            self.food_led.append(dep_details)
            self.food_tb += amount
            self.food_led_desc_only.append(description)
            self.food_tb_values.append(amount)
            
        # Repeating the process for this category:
        elif self.category == "Entertainment":
            self.ent_led.append(dep_details)
            self.ent_tb += amount
            self.ent_led_desc_only.append(description)
            self.ent_tb_values.append(amount)
        
        # repeating the process for this category:
        elif self.category == "Clothing":
            self.cl_led.append(dep_details)
            self.cl_tb += amount
            self.cl_led_desc_only.append(description)
            self.cl_tb_values.append(amount)
        
        # If the category is not mentioned, return False:
        else:
            return(False)
        
        # All details are placed to the overall ledger and balance:
        self.ledger.append(dep_details)
        self.total_balance += amount
        self.ledger_desc_only.append(description)
        self.total_balance_values.append(amount)
            
    # A withdraw method that is similar to the deposit method, but the amount passed 
    # in should be stored in the ledger as a negative number. If there are not enough funds, 
    # nothing should be added to the ledger. This method should return True if the withdrawal 
    # took place, and False otherwise.
    def withdraw(self, amount, description = None):
        
        # Same process with depositing, only deducted to the total balance:
        
        # Returning false when withdrawing a larger amount than what's deposited:
        if amount > 0:
            return(False)
        
        else:
        
            if description == None:
                description = ''
            else:
                description = description

            with_details = {"amount": amount, "description": description}

            if abs(amount) > self.total_balance:
                return(False)

            else:

                if self.category == "Food":
                    self.food_led.append(with_details)
                    self.food_tb += amount
                    self.food_led_desc_only.append(description)
                    self.food_tb_values.append(amount)
                    food_withlist.append(amount)
                    return(True)

                elif self.category == "Entertainment":
                    self.ent_led.append(with_details)
                    self.ent_tb += amount
                    self.ent_led_desc_only.append(description)
                    self.ent_tb_values.append(amount)
                    ent_withlist.append(amount)
                    return(True)

                elif self.category == "Clothing":
                    self.cl_led.append(with_details)
                    self.cl_tb += amount
                    self.cl_led_desc_only.append(description)
                    self.cl_tb_values.append(amount)
                    cl_withlist.append(amount)
                    return(True)

                else:
                    return(False)
    
    # A get_balance method that returns the current balance of the budget category 
    # based on the deposits and withdrawals that have occurred.
    def get_balance(self):
        return self.total_balance
    
    # A transfer method that accepts an amount and another budget category as arguments. 
    # The method should add a withdrawal with the amount and the description "Transfer to 
    # [Destination Budget Category]". The method should then add a deposit to the other 
    # budget category with the amount and the description "Transfer from [Source Budget Category]". 
    # If there are not enough funds, nothing should be added to either ledgers. 
    # This method should return True if the transfer took place, and False otherwise.
    def transfer(self, amount, dest_budcat):
        
        # Same process with depositing and withdrawing, the difference is that the value is -- 
        # -- transferred to another category:
        sour_budcat = self.category
        
        if amount > self.total_balance:
            return(False)
        
        else:
            transfer_to = "Transfer to "
            transfer_from = "Transfer from "
            dest_details = {"amount": -amount, "Transfer to": dest_budcat}
            sour_details = {"amount": amount, "Transfer from": sour_budcat}
            self.ledger.append(dest_details)
            self.ledger.append(sour_details)
            
            if sour_budcat == "Food":
                
                if dest_budcat == "Entertainment":
                    if amount <= self.food_tb:
                        self.ent_led.append(sour_details)
                        self.ent_tb += amount
                        self.ent_led_desc_only.append(transfer_from + sour_budcat)
                        self.ent_tb_values.append(amount)
                    else:
                        return(False)
                    
                elif dest_budcat == "Clothing":
                    if amount <= self.food_tb:
                        self.cl_led.append(sour_details)
                        self.cl_tb += amount
                        self.cl_led_desc_only.append(transfer_from + sour_budcat)
                        self.cl_tb_values.append(amount)
                    else:
                        return(False)
                else:
                    return(False)
                    
                self.food_led.append(dest_details)
                self.food_tb -= amount
                self.food_led_desc_only.append(transfer_to + dest_budcat)
                self.food_tb_values.append(-amount)

            elif sour_budcat == "Entertainment":
                
                if dest_budcat == "Food":
                    if amount <= self.ent_tb:
                        self.food_led.append(sour_details)
                        self.food_tb += amount
                        self.food_led_desc_only.append(transfer_from + sour_budcat)
                        self.food_tb_values.append(amount)
                    else:
                        return(False)
                    
                elif dest_budcat == "Clothing":
                    if amount <= self.ent_tb:
                        self.cl_led.append(sour_details)
                        self.cl_tb += amount
                        self.cl_led_desc_only.append(transfer_from + sour_budcat)
                        self.cl_tb_values.append(amount)
                    else:
                        return(False)
                else:
                    return(False)
                    
                self.ent_led.append(dest_details)
                self.ent_tb -= amount
                self.ent_led_desc_only.append(transfer_to + dest_budcat)
                self.ent_tb_values.append(-amount)
                
            elif sour_budcat == "Clothing":
                
                if dest_budcat == "Food":
                    if amount <= self.cl_tb:
                        self.food_led.append(sour_details)
                        self.food_tb += amount
                        self.food_led_desc_only.append(transfer_from + sour_budcat)
                        self.food_tb_values.append(amount)
                    else:
                        return(False)
                    
                elif dest_budcat == "Entertainment":
                    if amount <= self.cl_tb:
                        self.ent_led.append(sour_details)
                        self.ent_tb += amount
                        self.ent_led_desc_only.append(transfer_from + sour_budcat)
                        self.ent_tb_values.append(amount)
                    else:
                        return(False)
                else:
                    return(False)
                    
                self.cl_led.append(dest_details)
                self.cl_tb -= amount
                self.cl_led_desc_only.append(transfer_to + dest_budcat)
                self.cl_tb_values.append(-amount)
        
    def check_funds(self, amount):
        
        if self.category == "Food":
            if amount > self.food_tb:
                return(False)
            else:
                return(True)

        elif self.category == "Entertainment":
            if amount > self.ent_tb:
                return(False)
            else:
                return(True)
            
        elif self.category == "Clothing":
            if amount > self.cl_tb:
                return(False)
            else:
                return(True)
        else:
            return(False)
            
        
def create_spend_chart(categories):
    
    # Creating the string bar graph:
    sample_list = categories

    total_withdraw_food = 0
    total_withdraw_ent = 0
    total_withdraw_cl = 0
    total_withdraw_extra = 0

    print("Percentage spent by category")

    for value in food_withlist:
        total_withdraw_food += abs(value)
    for value in ent_withlist:
        total_withdraw_ent += abs(value)
    for value in cl_withlist:
        total_withdraw_cl += abs(value)
    for value in extra_withlist:
        total_withdraw_extra += abs(value)


    total_withdraw = total_withdraw_food + total_withdraw_ent+total_withdraw_cl+total_withdraw_extra

    percent_food = round(100*(total_withdraw_food / total_withdraw),-1)
    percent_ent = round(100*(total_withdraw_ent / total_withdraw),-1) 
    percent_cl = round(100*(total_withdraw_cl / total_withdraw),-1) 
    percent_extra = round(100*(total_withdraw_extra / total_withdraw),-1) 

    o_food = []
    o_ent = []
    o_cl = []
    o_extra = []

    n = 0
    for category in sample_list:
        if sample_list[n] == "Food":
            for i in range(0,11):
                if i*10 <= percent_food:
                    o_food.append("o") 
                else:
                    o_food.append(" ")

        elif sample_list[n] == "Entertainment":
            for i in range(0,11):
                if i*10 <= percent_ent:
                    o_ent.append("o") 
                else:
                    o_ent.append(" ")

        elif sample_list[n] == "Clothing":
            for i in range(0,11):
                if i*10 <= percent_cl:
                    o_cl.append("o") 
                else:
                    o_cl.append(" ")

        else: 
            for i in range(0,11):
                if i*10 <= percent_extra:
                    o_extra.append("o") 
                else:
                    o_extra.append(" ")
        n = n + 1

    line_1 = "100| " + o_food[10] + "  " + o_ent[10] + "  " + o_cl[10] + "  " + o_extra[10] 
    line_2 = " 90| " + o_food[9] + "  " + o_ent[9] + "  " + o_cl[9] + "  " + o_extra[9] 
    line_3 = " 80| " + o_food[8] + "  " + o_ent[8] + "  " + o_cl[8] + "  " + o_extra[8] 
    line_4 = " 70| " + o_food[7] + "  " + o_ent[7] + "  " + o_cl[7] + "  " + o_extra[7] 
    line_5 = " 60| " + o_food[6] + "  " + o_ent[6] + "  " + o_cl[6] + "  " + o_extra[6] 
    line_6 = " 50| " + o_food[5] + "  " + o_ent[5] + "  " + o_cl[5] + "  " + o_extra[5]
    line_7 = " 40| " + o_food[4] + "  " + o_ent[4] + "  " + o_cl[4] + "  " + o_extra[4] 
    line_8 = " 30| " + o_food[3] + "  " + o_ent[3] + "  " + o_cl[3] + "  " + o_extra[3] 
    line_9 = " 20| " + o_food[2] + "  " + o_ent[2] + "  " + o_cl[2] + "  " + o_extra[2] 
    line_10 = " 10| " + o_food[1] + "  " + o_ent[1] + "  " + o_cl[1] + "  " + o_extra[1] 
    line_11 = "  0| " + o_food[0] + "  " + o_ent[0] + "  " + o_cl[0] + "  " + o_extra[0] 

    if len(sample_list) > 3:
        line_12 = "    _____________"
    else:
        line_12 = "    __________"


    line_13 = "\n     F  E  C  A  " 
    line_14 = "     o  n  l  u  " 
    line_15 = "     o  t  o  t  " 
    line_16 = "     d  e  t  o  " 
    line_17 = "        r  h     "
    line_18 = "        t  i     "
    line_19 = "        a  n     "
    line_20 = "        i  g     "
    line_21 = "        n        "
    line_22 = "        m        "
    line_23 = "        e        "
    line_24 = "        n        "
    line_25 = "        t        "

    print(line_1)
    print(line_2)
    print(line_3)
    print(line_4)
    print(line_5)
    print(line_6)
    print(line_7)
    print(line_8)
    print(line_9)
    print(line_10)
    print(line_11)
    print(line_12)
    print(line_13)
    print(line_14)
    print(line_15)
    print(line_16)
    print(line_17)
    print(line_18)
    print(line_19)
    print(line_20)
    print(line_21)
    print(line_22)
    print(line_23)
    print(line_24)
    print(line_25)
```
