```python
# Setting the third argument as an optional one by assigning a default "None" value:
def add_time (start_time, duration_time, starting_day = None):
    
    # Splitting components of start_time
    split_start_time = start_time.split(" ")

    # Extracting the time component of start time:
    extract_start_time_number = split_start_time[0]

    # Extracting the AM/PM component of start time:
    extract_start_time_xM = split_start_time[1]

    # Splitting the time number and duration time further:
    split_time_number = extract_start_time_number.split(":")
    duration_time_number = duration_time.split(":")

    # Converting the time number into decimal, ending in an integer type:
    # If the start time is in PM, by real time it must have an additional 12 hrs.
    if extract_start_time_xM == "AM":
        # Special conditions for 12 as first digits when 'AM':
        if int(split_time_number[0]) == 12:
            decimal_time_number = (int(split_time_number[1]) / 60)
        else:
            decimal_time_number = int(split_time_number[0]) + (int(split_time_number[1]) / 60)        
    else: 
        # Special conditions for 12 as first digits when 'PM':
        if int(split_time_number[0]) == 12:
            decimal_time_number = int(split_time_number[0]) + (int(split_time_number[1]) / 60)
        else:
            decimal_time_number = int(split_time_number[0]) + (int(split_time_number[1]) / 60) + 12

    # Converting the duration time number into decimal, ending in an integer type:    
    decimal_duration_time_number = int(duration_time_number[0]) + (int(duration_time_number[1]) / 60)


    # Adding both start time and duration time to get the later time in integer:
    # Assigning the the values of the time into copies of the same variable for later use:
    decimal_time_later = decimal_time_number + decimal_duration_time_number
    decimal_time_later_copy1 = decimal_time_number + decimal_duration_time_number
    decimal_time_later_copy2 = decimal_time_number + decimal_duration_time_number


    # Defining all the variables:
    converted_dtl = ""
    index_day = 0
    days_ahead = 0
    number_week_days_ahead = 0
    every1_2hr = 0
    later_week_day = 0
    _multip = 0
    
    # Putting each week day in a list:
    week_days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

    # Setting the condition for the optional parameter:
    # Condition: The value of the optional parameter will remain "None" unless a value was assigned -- 
    # --  in the function:
    if starting_day != None:
        # Matching the assigned value to the list of values in the list:
        for week_day in week_days:
            # Using title function of strings to make the assigned value case insensitive:
            if starting_day.title() == week_day:
                # Breaking the for loop once a value is matched:   
                break 
            else:
                # Identifying the index or key value of the matched value:
                index_day = index_day + 1
    
    # If the optional argument is not supplied, the None value is retained.
    else: 
        starting_day = None

    # If the later time has a total greater than 24 hours:
    if decimal_time_later >= 24: 

        # Finding out how many whole days fit from the total time:
        while decimal_time_later_copy1 >= 24:
            decimal_time_later_copy1 = decimal_time_later_copy1 - 24
            
            # Storing the whole day count in this variable:
            days_ahead = days_ahead + 1

        # Finding out how many half days fit from the total time:
        while decimal_time_later_copy2 >= 12:
            decimal_time_later_copy2 = decimal_time_later_copy2 - 12
            
            # Storing the half day count in this variable:
            every1_2hr = every1_2hr + 1

        # A multiplier which returns how many weeks fit from the total time:
        multip_ = days_ahead // 7 

        # A condition which covers situations wherein the sum of the indeces and the whole day count -- 
        # -- exceeds the number of days in a week:
        if index_day + days_ahead > 7:
            
            # Resetting the count by decreasing the total days by the muptiplied 7:
            # Storing the value in a variable:
            number_week_days = (index_day + days_ahead) - (7*multip_)
            
            # Storing the value as an index of the list, in a variable:
            later_week_day = week_days[number_week_days]
            
        # The whole day count doesn't exceed the number of days in a week:    
        else:
            # Storing the value in a variable:
            number_week_days = index_day + days_ahead
            
            # Storing the value as an index of the list, in another variable:
            later_week_day = week_days[number_week_days]

        # The first copy now represents the value of the original, minus the whole days or half days.
        # If the value of the first copy is less than 1:
        if decimal_time_later_copy1 < 1:  

            # Assigning a variable for the first two digits of a time format:
            int_dtl = int(decimal_time_later_copy1) + 12
            
            # If the rounded value of the copy is less than 10, the 3rd and 4th digit of the time format --
            # -- must involve a '0' as its 3rd digit since the 4th digit is a single digit only.
            # 'AM' is automatically assigned as the copy has a value less than 1, implying the start of a day.
            if round(decimal_time_later_copy1*60) < 10:
                converted_dtl = str(int_dtl) + ":" +  "0" + str(round((decimal_time_later_copy1)*60)) + " " + "AM"
            
            # If the rounded value of the copy is less than 10, the 3rd and 4th digit of the time format --
            # -- this time, the condition now inputs both digits of the double digit minute time.
            # 'AM' is automatically assigned as the copy has a value less than 1, implying the start of a day.
            else:
                converted_dtl = str(int_dtl) + ":" +  str(round((decimal_time_later_copy1)*60)) + " " + "AM"

        # If the value of the first copy is less than 12:      
        elif decimal_time_later_copy1 < 12:
            
            # Assigning a variable for the first two digits of a time format:
            int_dtl_c1 = int(decimal_time_later_copy1)
            
            # If the rounded value of the copy is less than 10, the 3rd and 4th digit of the time format --
            # -- must involve a '0' as its 3rd digit since the 4th digit is a single digit only.
            # 'AM' is automatically assigned as the copy has a value less than 12, implying the mornings before 12 noon.
            if round((decimal_time_later_copy1 - int_dtl_c1)*60) < 10:
                converted_dtl = str(int_dtl_c1) + ":" +  "0" + str(round((decimal_time_later_copy1 - int_dtl_c1)*60)) + " " + "AM"
            
            # If the rounded value of the copy is less than 10, the 3rd and 4th digit of the time format --
            # -- this time, the condition now inputs both digits of the double digit minute time.
            # 'AM' is automatically assigned as the copy has a value less than 12, implying the mornings before 12 noon.
            else:
                converted_dtl = str(int_dtl_c1) + ":" +  str(round((decimal_time_later_copy1 - int_dtl_c1)*60)) + " " + "AM"

        # If the value of the first copy is more than 12:
        elif decimal_time_later_copy1 > 12:
            
            # Assigning a variable for the first two digits of a time format:
            int_dtl = int(decimal_time_later)
            
            # If the rounded value of the copy is less than 10, the 3rd and 4th digit of the time format --
            # -- must involve a '0' as its 3rd digit since the 4th digit is a single digit only.
            # 'PM' is automatically assigned as the copy has a value more than 12, from noon to the end of the day.
            if round((decimal_time_later - int_dtl)*60) < 10:
                converted_dtl = str(int(decimal_time_later_copy1)-12) + ":" + "0" + str(round((decimal_time_later - int_dtl)*60)) + " " + "PM"
            
            # If the rounded value of the copy is less than 10, the 3rd and 4th digit of the time format --
            # -- this time, the condition now inputs both digits of the double digit minute time.
            # 'PM' is automatically assigned as the copy has a value more than 12, from noon to the end of the day.
            else:
                converted_dtl = str(int(decimal_time_later_copy1)-12) + ":" + str(round((decimal_time_later - int_dtl)*60)) + " " + "PM"

    # If the later time has a total less than 24 hours:
    else:
        
        # Storing the index value of the week day input in a variable:
        number_week_days = index_day
        
        # Storing the value as an index of the list, in another variable:
        later_week_day = week_days[number_week_days]
    
        # If the value of the original total time is more than 12:
        if decimal_time_later >= 12:

            # Condition for the time format of the noon (e.g. 12:00 PM - 12:59 PM):
            if decimal_time_later < 13:

                # Assigning a variable for the first two digits of a time format:
                int_dtl = int(decimal_time_later)
                
                # If the rounded value of the original value is less than 10, the 3rd and 4th digit of --
                # -- the time format must involve a '0' as its 3rd digit since the 4th digit is a single digit only.
                # 'PM' is automatically assigned as the copy has a value range from 12 to 13 -- 
                # -- from noon to 12:59 PM.
                if round((decimal_time_later - int_dtl)*60) < 10:
                    converted_dtl = str(int_dtl) + ":" + "0" + str(round((decimal_time_later - int_dtl)*60)) + " " + "PM"

                # If the rounded value of the copy is less than 10, the 3rd and 4th digit of the time format --
                # -- this time, the condition now inputs both digits of the double digit minute time.
                # 'PM' is automatically assigned as the copy has a value range from 12 to 13 -- 
                # -- from noon to 1 PM.
                else:
                    converted_dtl = str(int_dtl) + ":" + str(round((decimal_time_later - int_dtl)*60)) + " " + "PM"

            # Condition for the time format of 1:00 PM onwards:
            else:
                
                # Assigning a variable for the first two digits of a time format:
                int_dtl = int(decimal_time_later)
                
                # If the rounded value of the original value is less than 10, the 3rd and 4th digit of --
                # -- the time format must involve a '0' as its 3rd digit since the 4th digit is a single digit only.
                # 'PM' is automatically assigned as the copy has a value range from 13 onwards or -- 
                # -- from 1 PM onwards.
                if round((decimal_time_later - int_dtl)*60) < 10:
                    converted_dtl = str(int_dtl - 12) + ":" + "0" + str(round((decimal_time_later - int_dtl)*60)) + " " + "PM"
                
                # If the rounded value of the original value is less than 10, the 3rd and 4th digit of --
                  # -- this time, the condition now inputs both digits of the double digit minute time.
                # 'PM' is automatically assigned as the copy has a value range from 13 onwards or -- 
                # -- from 1 PM onwards.
                else:
                    converted_dtl = str(int_dtl - 12) + ":" + str(round((decimal_time_later - int_dtl)*60)) + " " + "PM"

        # If the value of the original total time is less than 12:
        else:   
            
            # Assigning a variable for the first two digits of a time format:
            int_dtl = int(decimal_time_later)

            # If the rounded value of the original value is less than 10, the 3rd and 4th digit of --
            # -- the time format must involve a '0' as its 3rd digit since the 4th digit is a single digit only.
            # 'AM' is automatically assigned as the copy has a value range from midnight to noon.
            if round((decimal_time_later - int_dtl)*60) < 10:
                converted_dtl = str(int_dtl) + ":" + "0" + str(round((decimal_time_later - int_dtl)*60)) + " " + "AM"

            # If the rounded value of the original value is less than 10, the 3rd and 4th digit of --
            # -- this time, the condition now inputs both digits of the double digit minute time.
            # 'AM' is automatically assigned as the copy has a value range from midnight to noon.
            else:
                converted_dtl = str(int_dtl) + ":" + str(round((decimal_time_later - int_dtl)*60)) + " " + "AM"

    # Printing conditions for the function:
    # If the function doesn't have an assgined a value on the third optional argument:
    if starting_day == None:
        
        # If the the total time spans a more than day duration:
        if days_ahead > 1:
            print (converted_dtl + " (" + str(days_ahead) + " days later)") 
        
        # If the total time only spans up until the next day:
        elif days_ahead == 1:
            print (converted_dtl + " (next day)")
            
        # If the total time only spans within the day:
        else:
            print(converted_dtl)
            
    # If the function has an assgined a value on the third optional argument:
    elif starting_day != None:
        
        # If the the total time spans a more than day duration:
        if days_ahead > 1:
            print (converted_dtl + ", " + later_week_day + " (" + str(days_ahead) + " days later)")  
            
        # If the total time only spans up until the next day:
        elif days_ahead == 1:
            print(converted_dtl + ", " + later_week_day + " (next day)")
            
        # If the total time only spans within the day:
        else:
            print(converted_dtl + ", " + later_week_day)
            
```
