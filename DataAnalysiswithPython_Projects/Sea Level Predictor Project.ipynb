```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np

def draw_plot():
    # Read data from file
    df = pd.read_csv('epa-sea-level.csv')

    years_extended = np.arange(1880, 2080, 20)

    y = df["CSIRO Adjusted Sea Level"] 
    x = df['Year']

    slope, intercept, r_value, p_value, std_err = linregress(x, y)

    line = [slope*xi + intercept for xi in years_extended]

    fig, [ax1,ax2] = plt.subplots(figsize = (10,5),nrows = 1, ncols = 2)
    ax1.plot(years_extended, line, color = 'red', label="Fitting Line", linewidth=1)
    ax1.scatter(x, y, s = 5, marker = '.', label="Sample Point", color = 'dodgerblue')
    ax1.set_xlim(1880, 2060)
    ax1.set_ylim(-1,20)
    ax1.set_xticks(range(1880, 2080, 10))
    ax1.set_xticklabels(range(1880, 2080, 10),rotation = 90)
    ax1.set_xlabel('Year')
    ax1.set_ylabel("Sea Level (inches)")

    years_extended_new = np.arange(2000, 2080, 20)
    y_new = df.loc[df['Year'] >= 2000]["CSIRO Adjusted Sea Level"] 
    x_new = df.loc[df['Year']>= 2000]['Year']

    slope, intercept, r_value, p_value, std_err = linregress(x_new, y_new)

    line_new = [slope*xi + intercept for xi in years_extended_new]

    ax2.plot(years_extended_new, line_new, color = 'red', label="Fitting Line", linewidth=1)
    ax2.scatter(x_new, y_new, s = 5, marker = '.', label="Sample Point", color = 'dodgerblue')
    ax2.set_xlim(2000, 2060)
    ax2.set_ylim(-1,20)
    ax2.set_xticks(range(2000, 2080, 10), rotation = 90)
    ax2.set_xticklabels(range(2000, 2080, 10),rotation = 90)
    ax2.set_xlabel('Year')
    ax2.set_ylabel("Sea Level (inches)")

    plt.suptitle('Rise in Sea Level')
    # Save plot and return data for testing (DO NOT MODIFY)
    plt.savefig('sea_level_plot.png')
    return plt.gca()
```

```python
draw_plot()
```
